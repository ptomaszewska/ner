## Sprawdzanie zawodów

```python
def read_jobs(file):
    l = []
    with codecs.open(file , "r", encoding = 'UTF-8') as f:
        l = f.read().split('\n')
    return l
```

```python
folder = 'podzawody'
jobs = dict()
keys = read_jobs('zawody.txt')
for key in keys:
    jobs[key] = set(read_jobs(folder + '/' + key + '.txt'))
```

```python
import wikipedia
import morfeusz2
wikipedia.set_lang("pl")
```

```python
morf = morfeusz2.Morfeusz()
def exclude_job_for_person(person):
    # looks for one of jobs in wikipedia summary page
    global jobs
    profession = []
    global morf 
    try:
        page = wikipedia.page(person)
        l = []
        analysis = morf.analyse(page.summary)
        
        for i, j, interp in analysis:
            # Interp[0] - word in page.summary
            # Interp[1] - The base form of Interp[0] (with some tags or clear)
            #print(interp) #uncomment to see
            l.append(interp[0])
            l.append(interp[1])
        simple_words = set(l)
        for key in jobs.keys():
            if jobs[key].intersection(simple_words) != set():
                profession.append(key)
        if profession == []:
            profession.append('None')
    except:
        profession.append('None')
    page=None  #wcześniej był memory leak
    analysis=None #wcześniej był memory leak
    return profession
```

## Wstawianie tagów

```python
def create_tag (subset, index, doc, start, end, length_added):
    NER_name=subset.iloc[index,1]
    # if NER outputs only surname try to find whole name in corpora
    if " " not in NER_name:
        df=person.loc[person['surname'] == NER_name.lower()]
        if not df.empty:
            NER_name = str(df['name'])+str(df['surname'])

    if " " in NER_name:            
        NER_surname=NER_name.split()[1]
    else:
        NER_surname=NER_name
        
    #check if person exists in corpora to find its category
    if ((person.surname == NER_surname.lower()).any()):
        df=person.loc[person['surname'] == NER_surname.lower()]['category']
        category=str(df.values[0])
    else:
        #find category from wikipedia
        category=exclude_job_for_person(NER_name)
        category=category[0]

        if category == 'None':
            return None
        
    tag = ' <Entity name= "'+ doc[start+length_added: end+length_added] + '" type="person" category="'+category+'">'+NER_name+'<\Entity>'
    return tag
```

```python
def start_end_index(doc, index):
    return int(doc.iloc[index,0].split(":")[0]), int(doc.iloc[index,0].split(":")[1])
```

```python
def add_tag_to_doc (doc, subset): 
    length_added=0
    index=0
    #iterate over entities in a doc
    for i in range(subset.shape[0]):
        
        start, end=start_end_index(subset, i)
        tag=create_tag(subset, i, doc, start, end, length_added)
        if tag ==None:
            continue
        doc="".join((doc[:start+length_added],tag,doc[end+length_added:])) 
        length_added=length_added+len(tag)-(end-start)
    return doc
```
